```python

s = "Việc học tập là rất quan trọng, cho nên, cần học, học nữa, học mãi, học sinh."

processed_s = s.lower()
punctuation = ".,!?;:-"
for p in punctuation:
    processed_s = processed_s.replace(p, " ")

words = processed_s.split()

filtered_words = []
for word in words:
    if word.strip():  
        filtered_words.append(word)
words = filtered_words

print("Danh sách từ:", words)

word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print("Tần số từ:", word_count)


max_freq = 0
for freq in word_count.values():
    if freq > max_freq:
        max_freq = freq
print("Tần số cao nhất:", max_freq)

most_frequent_words = []
for word, freq in word_count.items():
    if freq == max_freq:
        most_frequent_words.append(word)

print("Từ xuất hiện nhiều nhất:", most_frequent_words)
```
