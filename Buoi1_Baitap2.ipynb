```python
"""Bài 2. Cho xâu s = "Việc học tập là rất quan trọng, cho nên, cần học, học nữa, học mãi".
1/ Xác định từ với tần số xuất hiện nhiều nhất, nếu có nhiều từ có cùng tần số nhiều nhất, in ra tất cả các từ đó (cùng với tần số của nó).
2/ Xác định độ dài của mỗi từ trong câu (không quan tâm đến lặp lại).
3/ Cho một bộ từ điển gồm các từ {học, nữa, học sinh, học bạ, mãi, học tập}, hãy xác định các từ xuất hiện trong bộ từ điển nói trên."""
```

```python
# Chuỗi cần phân tích
s = "Việc học tập là rất quan trọng, cho nên, cần học, học nữa, học mãi"

# 1. Hàm tìm từ xuất hiện nhiều nhất
def get_most_frequent_word(text):
    text = text.lower()
    words = []
    word = ""

    for char in text:
        if char.isalnum():  # chữ hoặc số
            word += char
        else:
            if word:
                words.append(word)
                word = ""
    if word:
        words.append(word)

    # Đếm tần suất
    word_count = {}
    for w in words:
        word_count[w] = word_count.get(w, 0) + 1

    max_freq = max(word_count.values())
    most_frequent_words = [w for w, freq in word_count.items() if freq == max_freq]
    return most_frequent_words

a = get_most_frequent_word(s)
print("Từ xuất hiện nhiều nhất là:", a)
```

```python
# Chuỗi cần phân tích
s = "Việc học tập là rất quan trọng, cho nên, cần học, học nữa, học mãi"

# 2. Hàm lấy độ dài từng từ và lọc trùng
def get_word_lengths(text):
    for char in ',.!?;:':
        text = text.replace(char, '')
    words = text.lower().split()

    # Lọc bỏ từ trùng
    unique_words = []
    for w in words:
        if w not in unique_words:
            unique_words.append(w)
    
    return [(word, len(word)) for word in unique_words]

lengths = get_word_lengths(s)
print("Độ dài của mỗi từ (không trùng) là:", lengths)
```

```python
# Chuỗi cần phân tích
s = "Việc học tập là rất quan trọng, cho nên, cần học, học nữa, học mãi"

# Bộ từ điển
dictionary = {"học", "nữa", "học sinh", "học bạ", "mãi", "học tập"}

def find_words_in_dictionary(text, dictionary):
    text_lower = text.lower()
    
    found_words = []
    for word in dictionary:
        # Xóa dấu câu trong text để so sánh dễ hơn
        clean_text = text_lower
        for char in ',.!?;:':
            clean_text = clean_text.replace(char, '')
        
        # Nếu từ/cụm từ có trong câu → thêm vào kết quả
        if word in clean_text.split() or word in clean_text:
            found_words.append(word)
    
    return found_words

# Gọi hàm
result = find_words_in_dictionary(s, dictionary)
for w in result:
    print(f"{w} xuất hiện trong bộ từ điển")
```

```python
s = "Việc học tập là rất quan trọng, cho nên, cần học, học nữa, học mãi"
# 3. Cách khác (không tối ưu) tìm từ xuất hiện nhiều nhất
def print_most_frequent_words_by_sorting(text):
    words = text.lower().split()
    words = [w.strip('.,!?;:') for w in words]
    words.sort()

    max_count = 0
    current_count = 1

    # Tìm max_count
    for i in range(1, len(words)):
        if words[i] == words[i - 1]:
            current_count += 1
        else:
            max_count = max(max_count, current_count)
            current_count = 1
    max_count = max(max_count, current_count)

    # In ra từ có max_count
    current_count = 1
    for i in range(1, len(words)):
        if words[i] == words[i - 1]:
            current_count += 1
        else:
            if current_count == max_count:
                print(f"Từ '{words[i - 1]}' xuất hiện {max_count} lần")
            current_count = 1
    if current_count == max_count:
        print(f"Từ '{words[-1]}' xuất hiện {max_count} lần")

print_most_frequent_words_by_sorting(s)
```
